**Q1.**

```python
import time
import matplotlib.pyplot as plt

def sum_iterative(n):
    total = 0
    for i in range(1, n+1):
        total += i
    return total

def sum_recursive(n):
    if n == 0:
        return 0
    return n + sum_recursive(n - 1)

# Measure time taken for different values of N
n_values = list(range(1, 1001, 100))
iterative_times = []
recursive_times = []

for n in n_values:
    start_time = time.time()
    sum_iterative(n)
    iterative_times.append(time.time() - start_time)

    start_time = time.time()
    sum_recursive(n)
    recursive_times.append(time.time() - start_time)

# Plotting
plt.plot(n_values, iterative_times, label='Iterative')
plt.plot(n_values, recursive_times, label='Recursive')
plt.xlabel('Value of N')
plt.ylabel('Time taken (seconds)')
plt.title('Time Complexity of Summing First N Natural Numbers')
plt.legend()
plt.show()
```

**Q2.**

```python
import time
import random
import matplotlib.pyplot as plt

# Generating random array
array = [random.randint(1, 1000) for _ in range(10000)]

def linear_search(arr, key):
    for i, val in enumerate(arr):
        if val == key:
            return i
    return -1

def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Measure time taken for 5 different searches
search_times_linear = []
search_times_binary = []

for _ in range(5):
    key = random.randint(1, 1000)

    start_time = time.time()
    linear_search(array, key)
    search_times_linear.append(time.time() - start_time)

    start_time = time.time()
    binary_search(array, key)
    search_times_binary.append(time.time() - start_time)

# Plotting
plt.plot(range(1, 6), search_times_linear, label='Linear Search')
plt.plot(range(1, 6), search_times_binary, label='Binary Search')
plt.xlabel('Search Number')
plt.ylabel('Time taken (seconds)')
plt.title('Time Complexity of Linear vs Binary Search')
plt.legend()
plt.show()
```

**Q3.**

```python
def string_to_int(s):
    if len(s) == 1:
        return int(s)
    return int(s[0]) * 10 ** (len(s) - 1) + string_to_int(s[1:])

# Example usage
print(string_to_int('13531'))  # Output: 13531
```

**Q4.**

```python
def reverse_string(s):
    if len(s) == 0:
        return s
    return reverse_string(s[1:]) + s[0]

# Example usage
print(reverse_string('pots&pans'))  # Output: snap&stop
```

**Q5.**

```python
def is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])

# Example usage
print(is_palindrome('racecar'))  # Output: True
print(is_palindrome('gohangasalamiimalasagnahog'))  # Output: True
```
